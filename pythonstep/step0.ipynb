## pythonは入っているか

Mac 及び メジャーなlinux distribution であれば pythonはデフォルトでinstall されています．
pythonの対話モードでの起動はターミナル(端末)で

    $ python
    Python 3.5.1 |Anaconda 2.5.0 (64-bit)| (default, Dec  7 2015, 11:16:01) 
    [GCC 4.4.7 20120313 (Red Hat 4.4.7-1)] on linux
    Type "help", "copyright", "credits" or "license" for more information.
    >>> 
    
とタイプすればpythonが起動し入力待ちになります． $マークは入力する必要はありません．

anacondaを使っていればipythonを使うほうが便利でしょう．

    $ ipython 
    Python 3.5.2 |Anaconda 2.5.0 (64-bit)| (default, Jul  2 2016, 17:53:06) 
    Type "copyright", "credits" or "license" for more information.

    In [1]:

pythonの終了は Ctrl + d (コントロールキーを押しながらdキーを押す)か exit, もしくは, exit() と入力してエンターキーを押して下さい．
なおipythonではTab補完がサポートされています．

## pythonのバージョン

pythonには2系，3系があり現在のメジャーなversionは3系です．
であるにも関わらず，macのデフォルトのpythonのversioは2.7です．
python2を使うべきかpython3を使うべきかは，意見が分かれる所ですが， 私は俄然3系をお勧めします． 
理由は数値計算を行う上で必要な最低限のライブラリーは3系でも現在充実しており， 
特段の理由が無い限り2系を使う必要が無いと思います．
とくに誤りを犯しやすい整数の割り算は3系の方が直感的な設計です．

python3系は

    >>> 1/3 
    0.3333333333333333

ですが，python2系は

    >>> 1/3
    0
    
となります．他にも細かな仕様の違いがありますが，3系から2系への移植は簡単なので，必要になれば2系に戻れば良いと思います．

## pythonのディストリビューション

pythonはversionの違いもありますが，様々な派生(ディストリビューション)があります．
詳しくは，[wiki](https://wiki.python.org/moin/PythonDistributions)に乗っていますが，その中でも科学計算を行う上で私は[anaconda](https://www.anaconda.com/download)の使用を勧めます．

主たる理由は，

* free
* windows, mac, linuxでpython2系及び3系をサポートしている
* ユーザー権限でinstallすることが可能でシステムのpythonと干渉しない
* 科学計算でよく用いる，numpy, scipy, matplotlibがデフォルトでinstallされている．
* パッケージ管理で利用するconda, pipがinstallされている
* intel math kernel library (MKL)による最適化が施されておりnumpy,scipy等の計算が爆速．詳しくは[ここを参照](https://www.anaconda.com/blog/developer-blog/anaconda-25-release-now-mkl-optimizations/)

等々です．intel自身もpythonのディストリビューションを配布(developer向け)していますが，特段の理由が無い限りanacondaをinstallするのがよいでしょう．

[anaconda](https://www.anaconda.com/download)のinstallはmac, linux では ダウンロードした*.shファイルを

```
$ bash Anaconda-なんちゃら.sh
```

として，質問に答えて行けばinstallが完了します．windowsの場合，exeファイルを実行すればinstallが完了します．
